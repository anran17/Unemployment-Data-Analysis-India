# Unemployment Analysis in India


## Dataset Description
    COLUMNS                         | DESCRIPTION
    Region                          | State or region
    Date                            | Month of data
    Estimated Unemployment Rate (%) | Percentage of unemployed people
    Estimated Employed	            | Number of employed people
    Labour Participation Rate	    | Workforce participation

## Import Libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset


```python
emp_data=pd.read_csv('Unemployment in India.xls')
```

## Data cleaning

```python
emp_data.columns=emp_data.columns.str.strip()
emp_data.columns
```

```python
emp_data['Date']=pd.to_datetime(emp_data['Date'])
```

## Exploratory Data Analysis

## Statistical Analysis

```python
emp_data.describe()
```

## Region Analysis

```python
emp_data.groupby('Region')['Estimated Labour Participation Rate (%)'].mean().sort_values(ascending=False).plot(kind='bar')
plt.title('Average Estimated Labour Participation Rate (%) by Region ')
plt.ylabel('Estimated Labour Participation Rate (%)')
plt.xlabel('Region')
```

```python
emp_data.groupby('Region')['Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False).plot(kind='bar')
plt.title('Average Estimated Unemployment Rate (%) by Region')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.show()
```

```python
emp_data.groupby('Region')['Estimated Employed'].mean().sort_values(ascending=False).plot(kind='bar')
plt.title('Average Estimated Employed by Region')
plt.xlabel('Region')
plt.ylabel('Estimated Employed')
plt.show()
```

## Urban vs Rural Analysis

```python
emp_data.groupby('Area')['Estimated Unemployment Rate (%)'].mean().plot(kind='bar')
plt.title('Area VS Rural (Estimated Unemployment Rate (%))')
plt.xlabel('Area')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.show
```

## Correlation Analysis Code

```python
corr=emp_data[['Estimated Labour Participation Rate (%)','Estimated Unemployment Rate (%)','Estimated Employed']].corr()
sns.heatmap(corr,annot=True, cmap='coolwarm')
plt.title("Correlation Between Variables")
plt.show()
```

## Line Chart – Unemployment Trend Over Time

```python
emp_data.groupby('Date')['Estimated Unemployment Rate (%)'].mean().plot()
plt.title('Unemployment Trend Over Time')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xlabel('Date')
plt.show()
```

```python
emp_data['Estimated Unemployment Rate (%)'].plot(kind='hist')
plt.xlabel('Estimated Unemployment Rate (%)')
plt.title('Distribution of Unemployment')
plt.show()
```
